```python
# importing networkx
import networkx as nx
# importing matplotlib.pyplot
import matplotlib.pyplot as plt
import xlrd

file ="C:\\Users\\Toshiba\\ijaz\\dataset\\dataset2.xls"


G=nx.Graph()

names=[]

book= xlrd.open_workbook(file)

sheet = book.sheet_by_index(0)

for row in range(sheet.nrows):
    data = sheet.row_slice(row)
    
    person1 = data[0].value
    person2 = data[1].value
    
    names.append((person1, person2))

G.add_edges_from(names)

nx.draw(G)
plt.show()
plt.savefig("C:\\Users\\Toshiba\\ijaz\\diagrams\\dataset2-graph.png")
```
